## Predict Abalone diameter 

### 1. Getting libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pdr
```

### 2. Prepare Data

```python
df = pd.read_csv('abalone.csv', names=['length','diameter'])
```

```python
length = np.array(df['length'])
```

```python
diameter = np.array(df['diameter'])
```

### 3. Functions for Linear Regression, Gradient Descent

```python
def hypothesis(w,b,x):
    h=w*x+b
    return h
def cost_function(Model,Output):
    c = (sum(pow((Model - Output), 2)))/(2*len(Model))
    return c
def GradientDecent(Input,Output,w, a):
    multi_wx = w*Input
    loss = multi_wx - Output
    gradient = np.sum(Input*loss) / len(Input)
    w=w-a*gradient
    return w
```

### 4. Linear Regression Model

```python
w=1.2 # 임의로 값 설정
b=0
Model = hypothesis(w,b,length)
```

```python
# 초기 임의로 구축한 선형 모델과 데이터 분포 확인
plt.plot(length,diameter,'or-')
plt.plot(length,Model,'ob-')
plt.show()
```

```python
a=0.3
parameter=[]
for i in range(200):
    w=GradientDecent(length,diameter,w, a)
    h=hypothesis(w,b,length)
    if (i%10)==0:
        print(w)
        parameter.append(w)
        plt.plot(length,diameter,'ob')
        plt.plot(length,h,'r')
        plt.show()
```

```python
# a값을 0.1으로 실행했을 때의 파라미터 w의 변화
para_array = np.array(parameter)
```

```python
plt.plot(para_array,'r')
plt.show()
```
